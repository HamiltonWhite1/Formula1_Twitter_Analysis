```python
import pandas as pd
```

```python
scrape_1 = pd.read_csv('Cleaned_Sets/first_scrape_clean.csv')
scrape_2 = pd.read_csv('Cleaned_Sets/second_scrape_clean.csv')
#skipped over a third in the web scrapes due to some testing
scrape_4 = pd.read_csv('Cleaned_Sets/fourth_scrape_clean.csv') 
scrape_5 = pd.read_csv('Cleaned_Sets/fifth_scrape_clean.csv')
scrape_6 = pd.read_csv('Cleaned_Sets/sixth_scrape_clean.csv')
scrape_7 = pd.read_csv('Cleaned_Sets/seventh_scrape_clean.csv')
scrape_8 = pd.read_csv('Cleaned_Sets/eighth_scrape_clean.csv')
scrape_9 = pd.read_csv('Cleaned_Sets/ninth_scrape_clean.csv')
scrape_10 = pd.read_csv('Cleaned_Sets/tenth_scrape_clean.csv')
```

```python
all_ten = pd.concat([scrape_1, scrape_2, scrape_4, scrape_5, scrape_6, scrape_7, scrape_8, scrape_9, scrape_10])
```

```python
all_ten.drop_duplicates(subset='full_text', inplace=True)
```

```python
for i in all_ten:
    print(i)
```

```python
count = 0
for tweet in all_ten['full_text']:
    if "max" in tweet:
        print(tweet + "\n")
        count+=1
print(count)
```
